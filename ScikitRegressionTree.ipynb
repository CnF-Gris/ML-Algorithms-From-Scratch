```python
import numpy
import seaborn as sns
from sklearn import datasets
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import GridSearchCV

from src.Models.Linear_Regression import Linear_Regression

[diabetes_X, diabetes_y] = datasets.load_diabetes(return_X_y=True)

diabetes_X = numpy.array(diabetes_X[:,2:])
diabetes_y = numpy.array(diabetes_y)

tree = DecisionTreeRegressor(
    min_samples_split=20
)

tree.fit(diabetes_X, diabetes_y)

predict_y = tree.predict(diabetes_X)
```

```python
from matplotlib import pyplot as plt
from sklearn.tree import plot_tree

fig = plt.figure(figsize=(60, 20))
plot_tree(tree, fontsize=10)
fig.show()
```

```python
sns.scatterplot(x=diabetes_X[:,1].ravel(), y=diabetes_y)
sns.lineplot(x=diabetes_X[:, 1].ravel(), y=predict_y)
```
