```python
from pandas import DataFrame
from sklearn import datasets




proc_data, y_true = datasets.make_blobs(
    n_samples=1000,
    n_features=2,
    centers=3,
    cluster_std=1,
    center_box=(-10.0, 10.0),
    shuffle=True,
)

DF = DataFrame(proc_data)
DF["Targets"] = y_true
```

```python
from sklearn.model_selection import GridSearchCV
from src.Models.K_Medoids import K_Medoids


model = K_Medoids()
search = GridSearchCV(model, {
    "clusters": [3, 5, 9]
})
model.fit(proc_data)
#clusters = search.predict(proc_data)
#print(search.cv_results_)

clusters = model.predict(proc_data)
DF["Pred"] = clusters
```

```python
import seaborn as sns

sns.scatterplot(DF, x=DF[0], y=DF[1], hue="Targets")
sns.scatterplot(DF, x=DF[0], y=DF[1], hue="Pred", palette=sns.color_palette("hls", 20))
```
